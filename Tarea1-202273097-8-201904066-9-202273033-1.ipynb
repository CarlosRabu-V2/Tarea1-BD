## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql

CREATE TABLE director (
    id_director INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INT NOT NULL,
    PRIMARY KEY (id_director)
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (
id_actor INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY ,
nombre VARCHAR(45) NOT NULL,
edad INT
);

```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_director INT NOT NULL,
    FOREIGN KEY (id_director) REFERENCES director (id_director),
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INT NOT NULL,
    fecha_publicacion DATE NOT NULL
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INT,
    id_pelicula INT,
    PRIMARY KEY (id_actor, id_pelicula),
    FOREIGN KEY (id_actor) REFERENCES Actor (id_actor),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula (id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_director INT,
    FOREIGN KEY(id_director) REFERENCES director(id_director),
    anio INT NOT NULL
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_actor INT,
    FOREIGN KEY(id_actor) REFERENCES actor(id_actor),
    anio INT NOT NULL
);

```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_pelicula INT,
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula),
    anio INT NOT NULL
);
```

**Solución:** _Selecciona los géneros de la tabla "pelicula" y cuenta cuántas veces aparece cada género. ### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**
|Nombre     |Cantidad de actores|
|-----------|-------------------|
|Oppenheimer|7                  |
|Poor Things|4                  |
|...        |...                |


**Solución:** _Para la obtención de esta consulta obtvimos el nombre e id_actores de cada pelicula de la tabla "pelicula", este ultimo se contó, usamos la tabla "actor_participa_en_pelicula" para relacionar peliculas con actores. Los resultados los agrupamos por nombre de pelicula. Se imprimen en orden decendiente las peliculas con mas de 2 actores._

```python
%%sql
SELECT p.nombre AS "Nombre", COUNT(ap.id_actor) AS "Cantidad de actores"
FROM pelicula p
JOIN actor_participa_en_pelicula ap ON p.id_pelicula = ap.id_pelicula
GROUP BY p.nombre
HAVING COUNT(ap.id_actor) >= 2
ORDER BY "Cantidad de actores" DESC;
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|
|------|------------------------|------------------------|
|Barbie|16                      |92                      |


**Solución:** _Para la solución de este problema se hacen 2 intersecciones entre pelicula, y acotr participa en pelicula, además de la intersección entre actor y el id de actor de la tabla actor participa en pelicula, el group by se utiliza para agrupar los resultados por nombre de pelicula y además se ocupa el MIN y MAX para poder obtener los resultados correctos._

```python
%%sql
SELECT
    p.nombre AS "Nombre",
    MIN(a.edad) AS "Edad del actor más joven",
    MAX(a.edad) AS "Edad del actor más viejo"
FROM
    pelicula p INNER JOIN actor_participa_en_pelicula ap 
    ON 
    p.id_pelicula = ap.id_pelicula INNER JOIN actor a 
    ON 
    ap.id_actor = a.id_actor
GROUP BY
    p.nombre;
```

### Consulta 7
**Año de cada premio ganado por la película de mayor duración.** (Si hay empate, elegir solamente uno).


**Ejemplo:**
|Año |
|----|
|2024|
|... |


**Solución:** _Selecciona aleatoriamente una película que tenga la duración máxima y que también haya ganado algún premio. Luego, extrae el año en que se otorgó el premio a esa película específica. Esto se logra mediante una serie de subconsultas anidadas que filtran y seleccionan las películas según ciertos criterios de duración y premios ganados, asegurando así que solo se incluyan películas elegibles. Finalmente, la consulta principal devuelve una lista de años únicos en los que se premió una película que cumple con estos requisitos._

```python
%%sql
SELECT DISTINCT anio AS "Año"
FROM (
    SELECT pelicula.*, premio_mejor_pelicula.anio
    FROM pelicula 
    INNER JOIN premio_mejor_pelicula ON pelicula.id_pelicula = premio_mejor_pelicula.id_pelicula
    WHERE pelicula.id_pelicula = (
        SELECT id_pelicula
        FROM (
            SELECT id_pelicula
            FROM pelicula
            WHERE duracion = (
                SELECT MAX(duracion)
                FROM pelicula
            )
            AND id_pelicula IN (
                SELECT id_pelicula
                FROM premio_mejor_pelicula
            )
            ORDER BY RANDOM()
            LIMIT 1
        ) AS pelicula_random
    )
) peliculas_premiadas_max_duración;
```

**Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**


**Ejemplo:**
|Nombre del actor|Cantidad de directores|Nombre de una película|
|----------------|----------------------|----------------------|
|Cillian Murphy  |29                    |Oppenheimer           |
|...             |...                   |...                   |


**Solución:** Para encontrar los actores que cumplen las condiciones obtuvimos cada actor de la tabla "actor_participa_en_pelicula" que relaciona actores de la tabla "actores" con peliculas de la tabla "pelicula" y contamos la candidad de directores diferentes para cada actor. Agrupamos los datos y si la "id_director" para cada actor es mayor a 1, se imprime. Finalmente se ordenó de manera decreciente. 

```python
%%sql
SELECT a.nombre AS "Nombre del actor",
       COUNT(DISTINCT p.id_director) AS "Cantidad de directores",
       MIN(p.nombre) AS "Nombre de una pelicula"
FROM actor_participa_en_pelicula ap
JOIN actor a ON ap.id_actor = a.id_actor
JOIN pelicula p ON ap.id_pelicula = p.id_pelicula
GROUP BY a.id_actor, a.nombre
HAVING COUNT(DISTINCT p.id_director) > 1
ORDER BY "Cantidad de directores" DESC;
```

**Solución:** Para obtener el resultado se hacen intersecciones entre la tabla actor_participa_en_pelicula y pelicula, tambíen entre el id, de actor_participa_en_pelicula y director, y tambien entre el id del direcotr de la tabla de pelicula y la tabla actor, además se agrega que se agrega que el actor haya participado en 1 a través del id del director, el distrinct hace que solo se seleccionen nombres únicos de actores, y el not exist se utiliza para verificar que solo hayan participado en 1 sola pelicula.

```python
%%sql

SELECT DISTINCT
    a.nombre AS "Nombre"
FROM
    actor_participa_en_pelicula ap INNER JOIN pelicula p 
    ON 
    ap.id_pelicula = p.id_pelicula INNER JOIN director d 
    ON 
    p.id_director = d.id_director INNER JOIN actor a 
    ON 
    ap.id_actor = a.id_actor
WHERE
    d.id_director = 1
    AND NOT EXISTS (
        SELECT 1
        FROM pelicula p2
        WHERE p2.id_director != 1
          AND p2.id_pelicula = p.id_pelicula
    );
```

**Solución:** _Utiliza varias uniones izquierdas para asegurar que se incluyan todos los actores y películas, incluso si no están asociados con premios de mejor director. Luego, cuenta el número de películas en las que cada director ha trabajado y el número de premios que han ganado. El filtro WHERE excluye los directores que no han ganado premios._

```python
%%sql
SELECT 
    a.nombre AS "Nombre del actor",
    pp.nombre AS "Nombre de la película",
    d.nombre AS "Nombre del director",
    COUNT(DISTINCT pp.id_pelicula) AS "Número de películas del director",
    COUNT(DISTINCT pd.id_premio) AS "Número de premios del director"
FROM 
    actor a
LEFT JOIN 
    actor_participa_en_pelicula ap ON a.id_actor = ap.id_actor
LEFT JOIN 
    pelicula pp ON ap.id_pelicula = pp.id_pelicula
LEFT JOIN 
    director d ON pp.id_director = d.id_director
LEFT JOIN 
    premio_mejor_director pd ON d.id_director = pd.id_director
WHERE 
    pd.id_premio IS NULL
GROUP BY 
    a.nombre, pp.nombre, d.nombre;

```
