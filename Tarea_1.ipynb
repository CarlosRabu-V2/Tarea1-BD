## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql

CREATE TABLE director (
    id_director INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INT NOT NULL,
    PRIMARY KEY (id_director)
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (
    id_actor INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER NOT NULL,
    PRIMARY KEY (id_actor)
);

```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INTEGER NOT NULL,
    fecha_publicacion DATE NOT NULL,
    PRIMARY KEY (id_pelicula),
    FOREIGN KEY (id_director) REFERENCES director (id_director)
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INTEGER,
    id_pelicula INTEGER,
    FOREIGN KEY (id_actor) REFERENCES Actor (id_actor),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula (id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio_dir INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER,
    PRIMARY KEY (id_premio_dir),
    FOREIGN KEY (id_director) REFERENCES pelicula (id_pelicula),
    anio INTEGER NOT NULL
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio_act INTEGER GENERATED ALWAYS AS IDENTITY,
    id_actor INTEGER,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio_act),
    FOREIGN KEY (id_actor) REFERENCES actor (id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (
    id_premio_pel INTEGER GENERATED ALWAYS AS IDENTITY,
    id_pelicula INTEGER,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio_pel),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula (id_pelicula)
);
```

### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**
|Nombre     |Cantidad de actores|
|-----------|-------------------|
|Oppenheimer|7                  |
|Poor Things|4                  |
|...        |...                |


**Solución:** Para la obtención de esta consulta obtvimos el nombre e id_actores de cada pelicula de la tabla "pelicula", este ultimo se contó, usamos la tabla "actor_participa_en_pelicula" para relacionar peliculas con actores. Los resultados los agrupamos por nombre de pelicula. Se imprimen en orden decendiente las peliculas con mas de 2 actores.

```python
%%sql
SELECT p.nombre AS "Nombre", COUNT(ap.id_actor) AS "Cantidad de actores"
FROM pelicula p
JOIN actor_participa_en_pelicula ap ON p.id_pelicula = ap.id_pelicula
GROUP BY p.nombre
HAVING COUNT(ap.id_actor) >= 2
ORDER BY "Cantidad de actores" DESC;
```

### Consulta 8
**Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**


**Ejemplo:**
|Nombre del actor|Cantidad de directores|Nombre de una película|
|----------------|----------------------|----------------------|
|Cillian Murphy  |29                    |Oppenheimer           |
|...             |...                   |...                   |


**Solución:** Para encontrar los actores que cumplen las condiciones obtuvimos cada actor de la tabla "actor_participa_en_pelicula" que relaciona actores de la tabla "actores" con peliculas de la tabla "pelicula" y contamos la candidad de directores diferentes para cada actor. Agrupamos los datos y si la "id_director" para cada actor es mayor a 1, se imprime. Finalmente se ordenó de manera decreciente. 

```python
%%sql
SELECT a.nombre AS "Nombre del actor",
       COUNT(DISTINCT p.id_director) AS "Cantidad de directores",
       MIN(p.nombre) AS "Nombre de una pelicula"
FROM actor_participa_en_pelicula ap
JOIN actor a ON ap.id_actor = a.id_actor
JOIN pelicula p ON ap.id_pelicula = p.id_pelicula
GROUP BY a.id_actor, a.nombre
HAVING COUNT(DISTINCT p.id_director) > 1
ORDER BY "Cantidad de directores" DESC;
```

**Nombre de todos los actores que han participado sólo en películas dirigidas por el director con id\_director 1.**


**Ejemplo:**
|Nombre      |

**Solución:** Para obtener el resultado se hacen intersecciones entre la tabla actor_participa_en_pelicula y pelicula, tambíen entre el id, de actor_participa_en_pelicula y director, y tambien entre el id del direcotr de la tabla de pelicula y la tabla actor, además se agrega que se agrega que el actor haya participado en 1 a través del id del director, el distrinct hace que solo se seleccionen nombres únicos de actores, y el not exist se utiliza para verificar que solo hayan participado en 1 sola pelicula.

```python
%%sql

SELECT DISTINCT
    a.nombre AS Nombre
FROM
    actor_participa_en_pelicula ap INNER JOIN pelicula p 
    ON 
    ap.id_pelicula = p.id_pelicula INNER JOIN director d 
    ON 
    p.id_director = d.id_director INNER JOIN actor a 
    ON 
    ap.id_actor = a.id_actor
WHERE
    d.id_director = 1
    AND NOT EXISTS (
        SELECT 1
        FROM pelicula p2
        WHERE p2.id_director != 1
          AND p2.id_pelicula = p.id_pelicula
    );
```
